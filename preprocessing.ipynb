```python
import pandas as pd
df = pd.read_csv('Student_Mental_Stress_and_Coping_Mechanisms.csv')
df.head()
```

```python
df.isnull().sum().sum()
```

```python
#membuat variabel df_missing, salinan df yang berisi missing values
import numpy as np

df_missing = df.copy()
np.random.seed(42)
for col in df_missing.columns:
    df_missing.loc[df_missing.sample(frac=0.1).index, col] = np.nan
```

```python
df_missing.isnull().sum()
```

```python
df_missing.head(10)
```

```python
#mengisi missing values dengan memberikan nilai paling dominan yang dimiliki atribut
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
df_missing = pd.DataFrame(imputer.fit_transform(df_missing),columns=df_missing.columns)
```

```python
df_missing.head(10)
```

```python
df_missing.isnull().sum()
```

```python
#menghapus kolom yang kurang relevan
df_missing = df_missing.drop(columns=['Student ID'])
```

```python
df_missing.columns
```

```python
df_missing.duplicated().sum()
```

```python
#memisahkan atribut dan target
x = df_missing.drop(columns=['Mental Stress Level'])
y = df_missing['Mental Stress Level']
```

```python
#mengubah atribut kategorikal menjadi numerik
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder',OneHotEncoder(),['Gender','Counseling Attendance','Stress Coping Mechanisms','Family Mental Health History', 'Medical Condition'])],remainder='passthrough')
x = np.array(ct.fit_transform(x))
```

```python
#membagi dataset, 20% untuk menguji dan 80% untuk melatih
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
ct.get_feature_names_out()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train[:,:3])
x_test_scaled = scaler.fit_transform(x_test[:,:3])
```
